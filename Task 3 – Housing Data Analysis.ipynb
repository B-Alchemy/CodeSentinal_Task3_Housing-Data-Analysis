**Import Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

**Load Dataset**

```python
df = pd.read_csv("kc_house_data.csv")
df.head()
```

**Correlation Matrix Heatmap**

```python
# Keep only numeric columns
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

**Top 5 Most Frequent Values in Selected Columns**

```python
for col in ['zipcode', 'bedrooms']:
    if col in df.columns:
        print(f"\nTop 5 most frequent values in {col}:")
        print(df[col].value_counts().head(5))
```

**Distribution Plots â€“ Histograms**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()
```

**Distribution Plots â€“ Boxplots**

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title("Price vs Bedrooms")
plt.show()
```

Summary â€“ Housing Data Analysis
- The correlation matrix shows that `sqft_living` and `grade` are most strongly correlated with price.  
- The most common number of bedrooms is 3, and certain zip codes dominate property listings.  
- Price distribution is right-skewed, with more lower-priced houses and fewer high-priced ones.  
- Houses with more bedrooms tend to have higher prices, but extremely large bedroom counts may lower price due to outdated design or non-standard layouts.
